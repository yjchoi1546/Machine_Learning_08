```python
import pandas as pd
import numpy as np

data = {
    'age':[25,27,np.nan,29,31,np.nan,35],
    'income':[50000, 54000, 58000, np.nan, 61000, 62000, np.nan]
}

df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()

# 데이터가 적을 때는 제거하면 안된다.
```

```python
df.dropna()
```

```python
df
```

```python
from sklearn.impute import SimpleImputer
pd.set_option('display.float_format', '{:.0f}'.format) # 포멧팅(소수점 제거해서 보여줌)
np.set_printoptions(suppress=True)#출력값 형태 변경


imputer = SimpleImputer(strategy='constant',fill_value=20)  #median : 중앙값 ,  most_frequent:최빈값, constant:고정값

result = imputer.fit_transform(df[['age','income']]) # fit은 학습을 시키는 것

df[['age','income']] = result
df[['age','income']]
```

```python
df[['age','income']]
# e+01 소수점을 하나 오른쪽으로 이동
```

```python
# 최빈값(most_frequent))으로 결측치 대체
imputer = SimpleImputer(strategy='most_frequent') #median : 중앙값, most_frequent:최빈값, constant:고정값


df[['age', 'income']] = imputer.fit_transform(df[['age','income']]) #fit 은 학습을 시키는 것이라고 이해

df[['age', 'income']]
```
